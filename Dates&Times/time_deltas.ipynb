```python
import pandas as pd
```

```python
#time deltas represent the amount or time, or difference between two datetimes
#.to_timedelta() method lets you use frequencies to return timedelta values 
#T=minute S=second D=Day W=week H=Hour pd.to_timedelta(5, unit="D")
```

```python
retail = pd.read_csv("../retail/retail.csv", parse_dates=["date"])
retail.head()
```

```python
retail.info(memory_usage="deep")
```

```python
retail = retail.assign(birthday=pd.to_datetime("1950-01-01"))

retail.head()
```

```python
retail = retail.assign(age=(retail["date"] - retail["birthday"]).dt.days / 365.25)
```

```python
retail.head()
```

```python
retail = retail.assign(retirement_date = retail["date"] + (pd.to_timedelta(365.25, unit="D") * 20))
retail.head
```
